```python
import pandas as pd
import sys
import plotly.express as px
```

```python
raw_data = pd.read_json("../../data/dynamodb_data.json")
```

```python
raw_data["publishedAt"] = pd.to_datetime(raw_data["publishedAt"])
raw_data.dtypes
```

```python
grouped_data = pd.DataFrame(raw_data.groupby("publishedAt").size()).reset_index()
grouped_data.columns = ["publishedAt", "count"]
# plot publishedAt
px.scatter(data_frame=grouped_data, y= "count", x="publishedAt")
```
